```python
data = pd.read_csv(".csv",encoding='cp949')
```

```python
data
```

```python
content_all = ''
for i in range(len(data['content'])):
    content_all = content_all + '' +data['content'].loc[i]
```

```python
data
```

```python
content_all
```

```python
import re
content_all = re.sub('[^--]',"",content_all)
content_all
```

```python
content_all = str(content_all)
content_all
```

```python
from konlpy.tag import Okt
okt = Okt()
data_txt = okt.nouns(content_all)
data_txt
```

```python
# from ckonlpy.tag import Twitter
# Twitter = Twitter()

# Twitter.add_dictionary('','Noun')
# Twitter.add_dictionary(' ','Noun')
from collections import Counter
count = Counter(data_txt)
count
rank_text = count.most_common()
```

```python
rank_text = dict(rank_text)
count_len = 10
temp_dic={}

for key, value in rank_text.items():
    if value > count_len:
        temp_dic[key] = value

rank_text = temp_dic

k_stopword = pd.read_csv("korean_stopword.csv")
k_stopword = list(k_stopword[''])
k_stopword.append('')
k_stopword.append('')
k_stopword.append('')
k_stopword.append('')
```

```python
temp_dic={}
for key, value in rank_text.items():
    if key not in k_stopword:
        temp_dic[key] = value
        
```

```python
temp_dic
```

```python
import matplotlib.font_manager as fm

#   
font_list = [font.name for font in fm.fontManager.ttflist]
font_list
```

```python
wordcloud = WordCloud(font_path = 'C:/Windows/Fonts/malgun.ttf', 
                      background_color='white',colormap = "Accent_r"
                     ).generate_from_frequencies(temp_dic)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
